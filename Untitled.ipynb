# Linear Algebra: Population interaction model

## Model explanation

The interaction of the two ethnic groups is modelled using a transition matrix, where each year’s population depends on the previous year:

$$x_{n+1} = \begin{bmatrix} 1.2 & 0.1 \\ 0.05 & 1.1 \end{bmatrix} x_n$$

**This means:**

Population X increases by 20% each year and is slightly influenced by population Y

Population Y increases by 10% each year and is slightly influenced by population X

## Figure interpretation

The population growth graph shows that both populations increase exponentially over time. This behaviour occurs because the dominant eigenvalue of the transition matrix is greater than 1, which indicates that the system is unstable but growing.

```python
#importing Library
import numpy as np
import matplotlib.pyplot as plt

# Transition matrix
A = np.array([[1.2, 0.1],
              [0.05, 1.1]])

# Initial population
x = np.array([100, 80])

years = 20
populations = []

for _ in range(years):
    populations.append(x)
    x = A @ x

populations = np.array(populations)

plt.plot(populations[:,0], label="Population X")
plt.plot(populations[:,1], label="Population Y")
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Growth of Two Ethnic Groups")
plt.legend()
plt.show()
```

# Calculus: Population growth of Paris

The population growth of Paris is modelled as:

$$P(t) = 1000e^{0.03t}$$

where **𝑡** represents time in years.

Growth rate analysis

The derivative of the population function gives the instantaneous growth rate:
$$P'(t) = 30e^{0.03t}$$
This shows that the population growth rate:

    Is always positive

    Increases exponentially over time

## Figure interpretation

The graph shows a smooth upward-curving exponential trend. The population grows slowly at first, but the growth accelerates in later years. The maximum growth rate occurs at the end of the 20-year period, indicating increasing population pressure over time.


```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 20, 100)
P = 1000 * np.exp(0.03 * t)

plt.plot(t, P)
plt.xlabel("Years")
plt.ylabel("Population")
plt.title("Population Growth of Paris Over 20 Years")
plt.show()
```

# Probability: Immigration events (Poisson distribution)

The number of random immigration events per year is modelled using a Poisson distribution with:

$$\lambda = 5$$
Expected value and variance

For a Poisson distribution:

    Expected value (mean) = 5

    Variance = 5

## Figure interpretation

The bar chart shows that:

    Most immigration events occur between 3 and 7 per year
    
    The probability peaks at 5 events per year
    
    Extreme values are unlikely

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

k = np.arange(0, 15)
prob = poisson.pmf(k, 5)

plt.bar(k, prob)
plt.xlabel("Number of Immigration Events")
plt.ylabel("Probability")
plt.title("Poisson Distribution (λ = 5)")
plt.show()
```

# District Population Analysis

Population data from eight districts in Paris were analysed to estimate the average district population. The sample mean was 1064.38 residents, with a standard deviation of 12.08, indicating low variation between districts. A 95% confidence interval for the true mean population was calculated as [1054.28, 1074.48].

These results suggest that district populations in Paris are consistent and evenly distributed. Although the sample size is small, the narrow confidence interval indicates a reliable estimate of the average population. This information can support effective population monitoring and informed resource planning across the city.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# District population data
data = np.array([1050, 1070, 1065, 1080, 1045, 1075, 1060, 1070])

# Sample size
n = len(data)

# Sample mean
mean = np.mean(data)

# Sample standard deviation (n-1)
std = np.std(data, ddof=1)

# 95% confidence interval using t-distribution
confidence_level = 0.95
df = n - 1
t_critical = stats.t.ppf((1 + confidence_level) / 2, df)

margin_of_error = t_critical * (std / np.sqrt(n))
ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error

# Print results
print("Sample Mean:", round(mean, 2))
print("Sample Standard Deviation:", round(std, 3))
print("95% Confidence Interval: [", round(ci_lower, 2), ",", round(ci_upper, 2), "]")

# Box plot with mean and confidence interval
plt.boxplot(data, vert=False)
plt.axvline(mean)
plt.axvline(ci_lower, linestyle='--')
plt.axvline(ci_upper, linestyle='--')

plt.xlabel("Population")
plt.title("District Population Distribution with Mean and 95% Confidence Interval")
plt.show()
```
